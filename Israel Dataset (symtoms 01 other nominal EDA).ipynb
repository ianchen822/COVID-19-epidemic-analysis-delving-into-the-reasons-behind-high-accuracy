```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 欄位資料分布視覺化

```python
df = pd.read_csv('Israel Dataset (syptoms 01 other nominal).csv')
df.tail()
```

```python
plt.figure(figsize = (16, 16))
column_names = list(df.columns.values)
for i in range(0, 9):
    if i == 8:
        plt.subplot(3, 3, i+1)
        x = list(set(df[column_names[i]]))
        y = [len(df[df[column_names[i]] == j]) for j in x]
        plt.bar(x, y)
    #         plt.ylabel('count')
        plt.title(column_names[i], fontsize = 10)
    elif i == 7:
        plt.subplot(3, 3, i+1)
        xl = np.arange(3)
        x = list(set(df[column_names[i]]))
        y1 = [len(df[(df[column_names[i]] == j) & (df['corona_result'] == 'positive')]) for j in x]
        y2 = [len(df[(df[column_names[i]] == j) & (df['corona_result'] == 'negative')]) for j in x]
        plt.bar(xl - 0.2, y1, width = 0.4, label = 'positive')
        plt.bar(xl + 0.2, y2, width = 0.4, label = 'negative')
    #         plt.ylabel('count')
        plt.xticks(xl, x)
        plt.legend()
        plt.title(column_names[i], fontsize = 10)
    else:
        plt.subplot(3, 3, i+1)
        xl = np.arange(2)
        x = list(set(df[column_names[i]]))
        y1 = [len(df[(df[column_names[i]] == j) & (df['corona_result'] == 'positive')]) for j in x]
        y2 = [len(df[(df[column_names[i]] == j) & (df['corona_result'] == 'negative')]) for j in x]
        plt.bar(xl - 0.2, y1, width = 0.4, label = 'positive')
        plt.bar(xl + 0.2, y2, width = 0.4, label = 'negative')
    #         plt.ylabel('count')
        plt.xticks(xl, x)
        plt.legend()
        plt.title(column_names[i], fontsize = 10)
plt.show()
```
