```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv('Israel Covid Dataset Preprocessed - Association.csv')
data.tail()
```

```python
data.replace(['T', 'F'], [1, 0], inplace = True)
data.tail()
```

```python
freq = apriori(data, min_support = 0.05, use_colnames = True)
result = association_rules(freq, metric = "confidence", min_threshold = 0.6)
```

```python
result.head(10)
```

```python
# 移除不重要欄位
data1 = data.drop(columns = ['test_indication_Other'])
freq = apriori(data1, min_support = 0.06, use_colnames = True)
result = association_rules(freq, metric = "confidence", min_threshold = 0.5)
result.head(10)
```

# 測試 min sup & min conf 組合

```python
minsups = []
minconfs = []

for i in np.arange(0.001, 0.1, 0.001):
    for j in np.arange(0.01, 0.9, 0.01):
        freq = apriori(data1, min_support = i, use_colnames = True)
        result = association_rules(freq, metric = "confidence", min_threshold = j)
        if len(result.index) >= 10:
            minsups.append(i)
            minconfs.append(j)
```

```python
maxsum = 0
for k in range(len(minsups)):
    sum = minsups[k] + minconfs[k]
    if sum > maxsum:
        maxsum = sum
        print('minsup:', minsups[k])
        print('minconf:', minconfs[k])
```

```python
for k in range(len(minsups)):
    print('minsup:', minsups[k])
    print('minconf:', minconfs[k])
```
